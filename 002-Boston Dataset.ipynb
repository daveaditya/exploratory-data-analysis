```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets

%matplotlib inline
```

```python
boston_data = datasets.load_boston()
```

```python
df = pd.DataFrame(boston_data['data'], columns=boston_data['feature_names'])
```

```python
df.head()
```

```python
df['TARGET'] = boston_data['target']
```

```python
print(boston_data['DESCR'])
```

```python
df.shape
```

There are 506 rows and 14 features in this dataset.

```python
df.info()
```

All the features are of type `float64`.

```python
df.describe()
```

```python
df[df.isna()].sum()
```

There are no missing values.

### Checking out the relation between Crime Rate and House Prices

```python
sns.scatterplot(x='CRIM', y='TARGET', data=df);
```

```python
sns.lmplot(x='CRIM', y='TARGET', data=df);
```

The is low `CRIM` in costly areas. Also, there is negative relation between `CRIM` and `TARGET`.

```python
sns.scatterplot(x='TARGET', y=np.arange(df.shape[0]), data=df, hue="CHAS");
```

### Checking the relation between `TAX` and `TARGET`

```python
sns.lmplot(x='TAX', y='TARGET', data=df);
```

As the `TAX` rate increases, the prices are low.

### Checking the relation between `ZN` and `TARGET`

```python
sns.lmplot(x='ZN', y='TARGET', data=df);
```

### Checking the relation between `NOX` and `TARGET`

```python
sns.lmplot(x='NOX', y='TARGET', data=df);
```

### Checking the relation between `AGE` and `TARGET`

```python
sns.lmplot(x='AGE', y='TARGET', data=df);
```

As expected, as the `AGE` increases the price of house i.e. the `TARGET` decreases.

### Checking the relation between `RM` and `TARGET`

```python
sns.lmplot(x='RM', y='TARGET', data=df);
```

As the number of rooms in a dwelling increases, the price also increases linearly.

### Checking the relation between `INDUS` and `TARGET`

```python
sns.lmplot(x='INDUS', y='TARGET', data=df);
```

As the area for non-retail increases, the target price of the houses also increases.

### Checking the relation between `DIS` and `TARGET`

```python
sns.lmplot(x='DIS', y='TARGET', data=df);
```

There is only slight increase in the target house price and hence no concrete relation between them.

## Correlation Matrix

```python
sns.heatmap(data=df.corr());
```

```python
positiv_corrs = df.corr()['TARGET'][df.corr()['TARGET'] > 0]
```

```python
sns.heatmap(data=pd.DataFrame(positiv_corrs));
```

There is high correlation between the target house price and the number of rooms per dwelling.
